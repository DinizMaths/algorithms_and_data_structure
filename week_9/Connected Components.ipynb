```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# connected componentes

# create a undirected graph
G = nx.Graph()

# create a branch of edges
G.add_edges_from([("a","e"),
                  ("e","d"),("e","f"),
                  ("d","f"),
                  ("f","h"),("f","g"),
                  ("h","i"),("h","b")])

# add a disconnected node
G.add_node("c")

# draw the graph
fig, ax = plt.subplots(1, 1,figsize=(10,8))
nx.draw_networkx(G, ax=ax,node_size=1000)
plt.axis("off")
plt.show()
```

```python
# G is connected or not?
nx.is_connected(G)
```

```python
# interact under all connected component of G
for component in nx.connected_components(G):
  print(component)
```

```python
# how many connected components has G?
nx.number_connected_components(G)
```

```python
# which connected component is a node N?
nx.node_connected_component(G, "a")
```

```python
# connected componentes

# create a directed graph
G = nx.DiGraph()

# create a branch of edges
G.add_edges_from([("a","e"),
                  ("e","d"),
                  ("d","f"),
                  ("f","h"),("f","g"),("f","e"),
                  ("h","b"),
                  ("i","h")])

# add a disconnected node
G.add_node("c")

# draw the graph
fig, ax = plt.subplots(1, 1,figsize=(10,8))
nx.draw_networkx(G, ax=ax,node_size=1000,pos=nx.circular_layout(G))
plt.axis("off")
plt.show()
```

```python
# is_connected is applied only to undirected graph, so is needed make a coversion before
nx.is_connected(nx.Graph(G))
```

```python
nx.is_strongly_connected(G)
```

```python
nx.is_weakly_connected(G)
```

```python
list(nx.weakly_connected_components(G))
```

```python
list(nx.strongly_connected_components(G))
```

```python
nx.number_strongly_connected_components(G)
```
